```python
import numpy as np

def monte_carlo_policy_evaluation(num_episodes):
    V = {s: 0 for s in range(12, 22)}

    for i in range(num_episodes):
        player_sum, dealer_card = 18,



        7
        episode = []
        while True:
            episode.append((player_sum, 'stand' if player_sum >= 18 else 'hit', 0))
            if player_sum >= 18:
                break
            player_sum += np.random.randint(1, 11)
            if player_sum > 21:
                episode.append((player_sum, 'bust', -1))
                return episode

        dealer_sum = dealer_card + np.random.randint(1, 11)
        while dealer_sum < 17:
            dealer_sum += np.random.randint(1, 11)
        if dealer_sum > 21 or player_sum > dealer_sum:
            episode[-1] = (episode[-1][0], episode[-1][1], 1)  # Player wins
        elif player_sum < dealer_sum:
            episode[-1] = (episode[-1][0], episode[-1][1], -1)  # Player loses
        else:
            episode[-1] = (episode[-1][0], episode[-1][1], 0)
        for state, i, reward in episode:
            V[state] += (reward - V[state]) / num_episodes
    return V

value_function = monte_carlo_policy_evaluation(1000)
print("Value function:", value_function)
```
