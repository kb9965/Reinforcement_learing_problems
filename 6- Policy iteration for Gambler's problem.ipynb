```python
def policy_iteration_gamblers_problem(target, p_h=0.4, theta=0.0001):
    V = [0] * (target + 1)
    pi = [0] * (target + 1)
    policy_stable = False

    while not policy_stable:
        while True:
            delta = 0
            for s in range(1, target):
                v = V[s]
                a = pi[s]
                win_state = s + a
                reward = 1 if win_state == target else 0
                V[s] = p_h * (reward + V[win_state]) + (1 - p_h) * V[s - a]
                delta = max(delta, abs(v - V[s]))
            if delta < theta:
                break

        policy_stable = True
        for s in range(1, target):
            old_action = pi[s]
            max_val = 0
            for a in range(1, min(s, target - s) + 1):
                win_state = s + a
                reward = 1 if win_state == target else 0
                val = p_h * (reward + V[win_state]) + (1 - p_h) * V[s - a]
                if val > max_val:
                    max_val = val
                    pi[s] = a
            if old_action != pi[s]:
                policy_stable = False

    return V, pi

target = 10


V, pi = policy_iteration_gamblers_problem(target)
print("Optimal value function:")
print(V)
print("Optimal policy:")
print(pi)
```
